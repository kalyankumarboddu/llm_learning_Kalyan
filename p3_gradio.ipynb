```python
import gradio as gr
import web_scraper as ws
from openai import OpenAI
```

```python
system_prompt="""
You are a professional assistant and good at extracting the website content and providing an complete accurate 
website summary
"""
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama_api_key = "Hardcode API secret key"
header = {
    "User-Agents":"Browser details"
}
```

```python
def fun_llm(url,model):
    selected_model = model
    ollama = OpenAI(base_url=OLLAMA_BASE_URL,api_key=ollama_api_key)
    response = ollama.chat.completions.create(
        model='llama3.2',
        messages=[
            {"role":"system","content":system_prompt},
            {"role":"user","content":ws.webpage_content(url,header)}
            ]
    )
    return response.choices[0].message.content
```

```python
def stream_llm (url,model):
    selected_model = model
    ollama = OpenAI(base_url=OLLAMA_BASE_URL,api_key=ollama_api_key)
    response = ollama.chat.completions.create(
        model='llama3.2',
        messages=[
            {"role":"system","content":system_prompt},
            {"role":"user","content":ws.webpage_content(url,header)}
            ],
        stream=True
    )
    # return response.choices[0].message.content
    result = ""
    for chunk in response:
        result += chunk.choices[0].delta.content
        yield result
```

```python
input_box = gr.Textbox(label="Input your Message",info="Enter website URL to be scrapped")
input_dropdown = gr.Dropdown(label="Model Selection",info="Select the required model",choices=['llama3.2','llama3.1'])
output_box = gr.Textbox(label="Output textbox",info="Ouptut displayed here",lines=10)
gr.Interface(
    fn=fun_llm,
    inputs=[input_box,input_dropdown],
    outputs=output_box,
    examples=[["htttps://google.com","llama3.2"],["htttps://huggingface.com","llama3.2"]]
).launch()
```

```python
input_box = gr.Textbox(label="Input your Message",info="Enter website URL to be scrapped")
input_dropdown = gr.Dropdown(label="Model Selection",info="Select the required model",choices=['llama3.2','llama3.1'])
output_box = gr.Markdown(label="Response")
gr.Interface(
    fn=stream_llm,
    inputs=[input_box,input_dropdown],
    outputs=output_box,
    examples=[["htttps://google.com","llama3.2"],["htttps://huggingface.com","llama3.2"]]
).launch(inbrowser=True)
```
