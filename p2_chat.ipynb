```python
from openai import OpenAI

OLLAMA_LOCAL_URL = "http://localhost:11434/v1"
ollama_bot = OpenAI(base_url=OLLAMA_LOCAL_URL,api_key="hardcoded_ollama_bot_api_key")
ollama_user = OpenAI(base_url=OLLAMA_LOCAL_URL,api_key="hardcoded_ollama_user_api_key")
```

```python
ollama_bot_system_prompt = """
you are a crazy assistant strickly responds with tuf and hard voice
"""

ollama_user_system_prompt = """
you are a lazy user always ask for silly questions
"""
```

```python
message_bot = [
    {"role": "system", "content": ollama_bot_system_prompt}
]
message_user = [
    {"role" : "system", "content" : ollama_user_system_prompt}
]

ollama_bot_prompt = ["Hi There"]
ollama_user_prompt = ["Hi"]
```

```python
def ollama_bot_chat(bot_message, user_message):
    for mess in zip(bot_message, user_message):
        message_bot.append({"role" : "assistant", "content" : mess[0]})
        message_bot.append({"role" : "user", "content" : mess[1]})
    response = ollama_bot.chat.completions.create(
        model = "llama3.2",
        messages = message_bot
    )
    return response.choices[0].message.content

```

```python
def ollama_user_chat(bot_message, user_message):
    for mess in zip(bot_message, user_message):
        message_user.append({"role": "assistant", "content": mess[1]})
        message_user.append({"role":"user", "content": mess[0]})
    response = ollama_user.chat.completions.create(
        model = "llama3.2",
        messages = message_user
    )
    return response.choices[0].message.content
```

```python
from IPython.display import Markdown


for i in range(5):
    display(Markdown("## Ollama Bot Response:"))
    ollama_bot_prompt[0] = ollama_bot_chat(ollama_bot_prompt,ollama_user_prompt)
    display(Markdown(ollama_bot_prompt[0]))

    display(Markdown("## Ollama User Response:"))
    ollama_user_prompt[0] = ollama_user_chat(ollama_bot_prompt,ollama_user_prompt)
    display(Markdown(ollama_user_prompt[0]))
```
