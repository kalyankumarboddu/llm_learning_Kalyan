```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"
```

```python
from openai import OpenAI

ollama = OpenAI(base_url=OLLAMA_BASE_URL,api_key='ollama_hardcode_any_value')

message = [
    {
        "role":"system", "content":"""Assume you are an analyst and will be given with an title name and list of url appending to the
        title, you can extract the information and those and share the summary what you feel"""
    },
    {
        "role":"user", "content" : "Sharing the title and list of urls" + links
    }
]
# response = ollama.chat.completions.create(model='llama3.2',messages=[{"role":"user", "content":"HI There!"}])

response = ollama.chat.completions.create(model="llama3.2", messages = message)
print(response.choices[0].message.content)
```
